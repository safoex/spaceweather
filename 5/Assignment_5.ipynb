```python
def rad_to_deg(rad):
    return rad*180/np.pi

def deg_to_rad(deg):
    return deg*np.pi/180
```

```python
def pixel_to_latlon(i,j, header):
    r = header['SOLAR_R']
    y_cor = i - header['CENTER_X']
    z_cor = header['CENTER_Y'] - j
    x_cor = np.sqrt(r**2 - y_cor**2 - z_cor**2)
    theta = np.arccos(z_cor / r)
    B = 90 - rad_to_deg(theta)
    phi = rad_to_deg(np.arctan2(y_cor, x_cor))
    return B, phi
```

```python
times_global = []
latitudes_global = []
longitudes_global = []

periods = ['June_2016', 'December_2010']
for period in periods:
    if 'June' in period:
        dataset_dir = june_directory
        files = june_files
        points = june_points
    else:
        dataset_dir = december_directory
        files = december_files
        points = december_points
        
    times = []
    latitudes = []
    longitudes = []
    for file, point in zip(files, points):
        filename = os.path.join(dataset_dir, file)
        header = fits.getheader(filename)
        lat, lon = (pixel_to_latlon(point[0], point[1], header))
        time = header['DATE_OBS']
        time =  datetime.strptime(time, '%Y-%m-%dT%H:%M:%SZ')
        times.append(time)
        latitudes.append(lat)
        longitudes.append(lon)
    times_global.append(times)
    latitudes_global.append(latitudes)
    longitudes_global.append(longitudes)
        
    fig = plt.figure(figsize=(5, 5))
    ax1 = fig.add_subplot(111)
    ax1.plot(times, latitudes, label='Latitude')
    ax1.plot(times, longitudes, label='Longitude')
    ax1.set_xlabel('Time')
    ax1.set_ylabel('Degrees')
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y\n%m\n%d'))
    ax1.legend()
    ax1.set_title('Latitudes and longitudes from time, {}'.format(period))
    plt.show()
    
    plt.figure(figsize=(5, 5))
    plt.plot(longitudes, latitudes)
    plt.xlabel('Degrees, longitude')
    plt.ylabel('Degrees, latitude')
    plt.title('Latitudes from longitudes, {}'.format(period))
    plt.show()
```
