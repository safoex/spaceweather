```python
import astropy.io.fits as fits
import scipy as scp
import numpy as np
import matplotlib.pyplot as plt
import itertools
import collections
def iterer(*args):
    return itertools.product(*[x_ if isinstance(x_,collections.Iterable) else range(x_) for x_ in args])
import os
import matplotlib.dates as mdates
from datetime import datetime, timedelta
%matplotlib inline
```

# First we have to analyze sunspot movements

```python
june_directory = '2016_June/'
june_files = sorted(os.listdir(june_directory))
june_points = [[120, 1110], [185, 1130], [285, 1130], [410, 1130], [620, 1140], [795, 1140], [1000, 1145], [1210, 1145], [1410, 1145], [1610, 1145], [1857, 1135], [1930, 1125]]
```

```python
december_directory = '2010_December/'
december_files = sorted(os.listdir(december_directory))
december_points = [[430, 550], [550, 540], [695, 535], [890, 535], [1030, 535], [1250, 525], [1395, 525], [1540, 525]]
```

```python
def pixel_to_latlon(i,j, header):
    r = header['SOLAR_R']
    y_cor = i - header['CENTER_X']
    z_cor = header['CENTER_Y'] - j
    x_cor = np.sqrt(r**2 - y_cor**2 - z_cor**2)
    theta = np.arccos(z_cor / r)
    B = 90 - theta*180/np.pi
    phi = np.arctan2(y_cor, x_cor)*180/np.pi
    return B, phi
```

```python
periods = [(june_directory, june_files, june_points), (december_directory, december_files, december_points)]
def get_header(dataset_dir, file):
    filename = os.path.join(dataset_dir, file)
    return fits.getheader(filename)
        
latitudes  = [[pixel_to_latlon(point[0], point[1], get_header(dataset_dir, file))[0] for file, point in zip(files, points)] for dataset_dir, files, points in periods] 
longitudes = [[pixel_to_latlon(point[0], point[1], get_header(dataset_dir, file))[1] for file, point in zip(files, points)] for dataset_dir, files, points in periods] 
times = [[datetime.strptime(get_header(dataset_dir, file)['DATE_OBS'], '%Y-%m-%dT%H:%M:%SZ') for file, point in zip(files, points)] for dataset_dir, files, points in periods]
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,10), dpi=300)
for i in range(2):
    ax = axes[i][0]
    ax.plot(times[i], latitudes[i], label='Latitude')
    ax.plot(times[i], longitudes[i], label='Longitude')
    #ax.set_xlabel('Time')
    ax.set_ylabel('deg')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y\n%m\n%d'))
    ax.legend()
    #ax.set_title('Latitudes and longitudes from time, {}'.format(periods[i][0][:-1]))
    ax = axes[i][1]
    ax.plot(longitudes[i], latitudes[i])
    ax.set_xlabel('deg, longitude')
    ax.set_ylabel('deg, latitude')
    #ax.set_title('Latitudes from longitudes, {}'.format(periods[i][0][:-1]))
plt.tight_layout()
```

Sun has a rotation not only at one axe in Earth coordinate system
